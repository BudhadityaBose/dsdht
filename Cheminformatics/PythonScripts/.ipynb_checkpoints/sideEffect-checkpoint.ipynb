```python
import pandas
import csv
import gzip
import collections
```

```python
base_url = 'http://sideeffects.embl.de/media/download/'
filenames = [
    'meddra_all_indications.tsv.gz',
    'meddra_all_se.tsv.gz',
    'meddra_freq.tsv.gz']
for filename in filenames:
    ! wget --no-verbose --timestamping --directory-prefix download {base_url}/{filename}
```

```python
drugbank_df = pandas.read_table('data/drugbank.tsv')[['drugbank_id', 'name']].rename(columns={'name': 'drugbank_name'})
dbpc_map_df = pandas.read_table('data/pubchem-mapping.tsv')

dbpc_map_df.head(5)
```

# medra freq

```python
columns = [
    'stitch_id_flat',
    'stitch_id_sterio',
    'umls_cui_from_label',
    'placebo',
    'frequency',
    'lower',
    'upper',
    'meddra_type',
    'umls_cui_from_meddra',
    'side_effect_name',
]
freq_df = pandas.read_table(gzip.open('download/meddra_freq.tsv.gz'), names=columns)
freq_df.head(2)
```

# meddra_all

```python
def stitch_flat_to_pubchem(cid):
    assert cid.startswith('CID')
    return int(cid[3:]) - 1e8

def stitch_stereo_to_pubchem(cid):
    assert cid.startswith('CID')
    return int(
cid[3:])

columns = [
    'stitch_id_flat',
    'stitch_id_sterio',
    'umls_cui_from_label',
    'meddra_type',
    'umls_cui_from_meddra',
    'side_effect_name',
]
se_df = pandas.read_table(gzip.open('download/meddra_all_se.tsv.gz'), names=columns)
se_df['pubchem_cid'] = se_df.stitch_id_sterio.map(stitch_stereo_to_pubchem)
se_df = dbpc_map_df.merge(se_df)
se_df.head(2)
```

```python
se_df = se_df[['drugbank_id', 'umls_cui_from_meddra', 'side_effect_name']]
se_df = se_df.dropna()
se_df = se_df.drop_duplicates(['drugbank_id', 'umls_cui_from_meddra'])
se_df = drugbank_df.merge(se_df)
se_df = se_df.sort(['drugbank_id'])
se_df.head(5)
```

#  meddra_all_indications

```python
columns = [
    'stitch_id_flat',
    'umls_cui_from_label',
    'method',
    'concept_name',
    'meddra_type',
    'umls_cui_from_meddra',
    'meddra_name',
]
indication_df = pandas.read_table(gzip.open('download/meddra_all_indications.tsv.gz'), names=columns)
indication_df['pubchem_cid'] = indication_df.stitch_id_flat.map(stitch_flat_to_pubchem)
indication_df = drugbank_df.merge(dbpc_map_df.merge(indication_df))
indication_df = indication_df.query("meddra_type == 'PT'")
indication_df.head(2)
```
